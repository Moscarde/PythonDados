```python
url_bc_pix = '''https://olinda.bcb.gov.br/olinda/servico/Pix_DadosAbertos/versao/v1/odata/TransacoesPixPorMunicipio(DataBase=@DataBase)?@DataBase='202011'&$top=10000&$format=json'''

pix_json = requests.get(url_bc_pix).json()
df_pix = pd.DataFrame(pix_json['value'])
```

## Dinheiro em Circulação - v1

```python
pular = 0

lista = []
while True:

    url_dinheiro_circulação = f'''https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=10000&$skip={pular}&$format=json'''
    req_json = requests.get(url_dinheiro_circulação).json()['value']

    
    lista.append(pd.DataFrame(req_json))
    pular += 10000

    print('cc')
    if len(req_json) != 10000:
        
        break

lista
```

```python
df_dinheiro = pd.concat(lista)
df_dinheiro
```

```python
# Moedas comemorativas
pular = 0
lista_requests = []
while True:
    url_moedas = f'''https://olinda.bcb.gov.br/olinda/servico/mecir_moedas_comemorativas/versao/v1/odata/informacoes_diarias?$top=10000&$skip={pular}&$format=json'''
    moedas_json = requests.get(url_moedas).json()['value']

    if len(moedas_json) < 1:
        break
    
    df_req = pd.DataFrame(moedas_json)
    lista_requests.append(df_req)
    pular += 10000

df_moedas = pd.concat(lista_requests)
    
    
```
