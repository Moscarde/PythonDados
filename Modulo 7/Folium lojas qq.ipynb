```python
import folium
from folium.plugins import MarkerCluster, FeatureGroupSubGroup
import pandas as pd
from pprintpp import pprint
import time

%reload_ext dotenv
import googlemaps
import os

chave_api = os.getenv("API_KEY")
client = googlemaps.Client(chave_api)
```

```python
df_qq = pd.read_csv("lojas_qq_filtrado.csv")
coordenadas_qq = df_qq[["latitude", "longitude"]].values
enderecos = df_qq['endereco']
df_qq
```

```python
lista_estados = []
endereco_erro = []

for i, endereco in enumerate(enderecos):
    response = client.geocode(endereco + ', BR')

    if len(response) == 0:
        endereco_erro.append([response, endereco])

    else:
        for address_type in response[0]['address_components']:
            if 'administrative_area_level_1' in address_type['types']:
                endereco_estado = [endereco, address_type['short_name']]
                lista_estados.append(endereco_estado)

        print(len(enderecos) - i)

    time.sleep(.3)
print(len(lista_estados))
```

```python
lista_dfs = []
for endereco in lista_estados:
    df = pd.DataFrame({'endereco': endereco[0], 'UF': endereco[1]}, index= [0])
    lista_dfs.append(df)

df_uf = pd.concat(lista_dfs)
df_uf
```

```python
df_qq_merged = pd.merge(left=df_qq, right=df_uf, how='left', left_on=['endereco'], right_on=['endereco'])
df_qq_merged = df_qq_merged.dropna()
df_qq_merged
```

```python
mapa = folium.Map(location=[-25, -50], zoom_start=5)

marker_cluster = MarkerCluster().add_to(mapa)


for i in range(len(df_qq_merged)):

    folium.Marker(
        location=[df_qq_merged.iloc[i]["latitude"], df_qq_merged.iloc[i]["longitude"]],
        popup=df_qq_merged.iloc[i]["endereco"],
        icon= None
    ).add_to(marker_cluster)

mapa
```
