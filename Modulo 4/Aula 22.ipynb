```python
# Merge x Join

dados_roe = pd.DataFrame({'ticker': ['WEGE3', 'VALE3', 'PCAR3', 'ELET3', 'MGLU3'],
'ROE': [20, 9, 19, 3, 5]})
dados_roic = pd.DataFrame({'codigo': ['WEGE3', 'VALE3', 'PCAR3', 'ELET3', 'VIIA3'],
'ROIC': [25, 6, 13, 1, 2]})

indicadores = pd.merge(dados_roe, dados_roic, how='inner', left_on=['ticker'], right_on=['codigo'])
indicadores
```

Merge Methods - grupos
- left ~> usa chaves da esquerda apenas
- right ~> usa chaves da direita apenas
- inner ~> usa apenas chaves que interseccionam
- outer ~> usa todas as chaves completando com NaN as que nao estao aparecem em outro df

```python
print('outer')
print(pd.merge(dados_roe, dados_roic, how='outer', left_on=['ticker'], right_on=['codigo']))
print('left')
print(pd.merge(dados_roe, dados_roic, how='left', left_on=['ticker'], right_on=['codigo']))
```

```python
dados_cotacoes = yf.download('ABEV3.SA')
dados_pl = pd.DataFrame({
    'ticker': ['ABEV3'] * len(dados_cotacoes),
    'PL': np.random.randint(10, 20, len(dados_cotacoes))
    },
    index = dados_cotacoes.index
    )
print(dados_cotacoes)
print(dados_pl)
```

```python
dados_finais = dados_cotacoes.join(dados_pl)
dados_finais.head(5)
```

```python
#removendo a coluna ticker
dados_finais = dados_finais.drop('ticker', axis=1)
# dados_finais.drop('ticker', axis=1, inplace = True) inplace = True realiza a operacao no proprio df (sem retorno)
dados_finais.head(5)
```
