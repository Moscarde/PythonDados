## Código dos paises segue o ISO 3166
[https://en.wikipedia.org/wiki/List_of_ISO_3166_country_codes](https://en.wikipedia.org/wiki/List_of_ISO_3166_country_codes)

```python
#Calculo com PIB - Multiplos paises
# GDP per capita (constant 2015 US$)
pib_per_capta = wb.download(
    indicator='NY.GDP.PCAP.KD',
    country = ['MX', 'BRA', 'USA', 'ARG', 'UKR'],
    start=1980,
    end= 2023)
print(pib_per_capta)
```

```python
aumento_em_40_anos = pib_per_capta.groupby('country').first() / pib_per_capta.groupby('country').last() - 1

aumento_em_40_anos
```

```python
#62 CO2 emissions kt ~~ 1990 agregado no mundo
matches = wb.search('CO2 emissions.*(kt)')
matches
```

```python
co2_emissions = wb.download(
    indicator="EN.ATM.CO2E.KT", 
    start=1990, 
    end=2023, 
    country=['1W']
    )
co2_emissions
```

```python
co2_emissions = co2_emissions.dropna()
co2_emissions = co2_emissions.droplevel(level= 0)
co2_emissions = co2_emissions.sort_index()
```

```python
co2_emissions.plot()
```
