```python
df_cambio = pd.DataFrame({
    'moeda': ['dolar/real', 'libra/real'] * 4,
    'price': [5.24, 4.15, 5.26, 4.23, 5.55, 4.02, 5.76, 3.98],
    'derivativo': list(['DOLFT3', 'LIBFT3']) * 4},
    index = sorted(list(pd.date_range('1/5/2014', periods = 4)) * 2)
)

df_cambio
```

```python
#pegando dados uncios de uma coluna
ativos = df_cambio['derivativo'].unique()
ativos
```

```python
for ativo in ativos:
    df_cambio_filtrado = df_cambio[df_cambio['derivativo'] == ativo]

    df_cambio_filtrado['price'] = df_cambio_filtrado['price'] / 100
    df_cambio_filtrado['moeda'] = df_cambio_filtrado['moeda'].str.replace('real', 'banana')
    print(df_cambio_filtrado)

#geralmente groupby vai resolver isso melhor...
```

```python
dados_mercado = pd.DataFrame({
    'Nome': ['Lojas Quero Quero', 'Alpargatas', 'Alpargatas', 'Vale', 'C&A'],
    'Codigos': ['LJQQ3', 'ALPA3', 'ALPA4', 'VALE3', 'CEAB3'],
    'Cotacoes': [20, 25, 30, 21, 23],
    'Volume': [1000, 2000, 3500, 10000, 50000]
})
dados_mercado
```

### Desafio


```python

#df base
dict_desafio = {
    "tickers" : ["WEGE3", "PETR3", "VALE3", "PETR4", "LREN3"],
    "cotacoes": [20, 30, 40, 12, 35],
    "nomes": ['Weg', 'Petro', 'Vale', 'Petro', 'Lojas Renner'],
    "preco_sobre_lucro": [25, 6, 12, 7, 25],
    "volume": [5000, 1000, 4000, 7000, 1200]
    }

df = pd.DataFrame(dict_desafio, index= [1, 2, 3, 4, 5])
df = df.set_index("tickers")
df.columns = ['preco', 'nomes', 'preco_sobre_lucro', 'volume']
df['lucro_por_acao'] = df['preco'] / df['preco_sobre_lucro']
dolar = 5.25
df['preco_em_dolar'] = df['preco'] / dolar
df['rank_lpa'] = df['lucro_por_acao'].rank(method = 'dense')
df
```

```python
df.drop_duplicates(subset='nomes')
```

```python
# mantendo o tick mais negociado da petro
df.sort_values(by= 'volume', ascending=False).drop_duplicates(subset= 'nomes')
```
