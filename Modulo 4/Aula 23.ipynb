# Pandas

```python
import pandas as pd
import yfinance as yf
```

```python
dados_weg = yf.download('WEGE3.SA')['Adj Close']
dados_weg
```

```python
#fechamento mensal
dados_weg_mensal = dados_weg.resample('M').last()
dados_weg_mensal
```

```python
#maximo anual
dados_weg_anual = dados_weg.resample('Y').max()
dados_weg_anual
```

```python
#retorno anual
dados_weg_anual.pct_change()
```
