```python
valores = np.random.rand(10)
valores
```

```python
#fatiando valores baseado em %
divisoes = [0, 0.25, 0.5, 0.75, 1]
valores_fatiados = pd.cut(valores, divisoes)
valores_fatiados
#retorna em qual campo de divisioes está cada valor
```

```python
numero_de_amostras_por_categoria = pd.value_counts(valores_fatiados)
numero_de_amostras_por_categoria
#conta quantos valores se repetem em cada divisao
```

```python
#transformar a amostra em variaveis dummys

df_dummies = pd.get_dummies(valores_fatiados)
df_dummies
```
