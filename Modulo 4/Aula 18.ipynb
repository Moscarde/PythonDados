```python
#substituindo dados numa coluna com .replace

dados_euca3 = pd.read_excel('dados_euca3.xlsx')
dados_euca3
```

```python
dados_euca3['EUCA3'] = dados_euca3['EUCA3'].replace('nd', pd.NA)
print(dados_euca3)

dados_euca3['EUCA3'] = dados_euca3['EUCA3'].replace(pd.NA, 0)
print(dados_euca3)
```

```python
#cuidado com o replace em strings, se desejar substituir uma parte da string
#o metodo str deve ser chamado antes do replace

df_strings = pd.Series(['Ordem: Compra', 'Ordem: Hold'])
print(df_strings.replace('Compra', 'Venda'))

df_strings = df_strings.str.replace(': Compra', '~> Venda')

df_strings
```

```python
numeros = pd.Series([20, 30, pd.NA, 25, 22, pd.NA, 17])
numeros
```

```python
df_numeros = pd.DataFrame({'numeros': numeros})
# numeros.replace(pd.NA, 0)
print(df_numeros['numeros'].replace(pd.NA, 0))
print(df_numeros.dropna())
```

```python
#metodo aula, calculando retorno diario
import numpy as np

print(numeros)
numeros = numeros.fillna(0)

retornos = []

for i, dado in enumerate(numeros):
    if i == 0:
        retornos.append(pd.NA)
    
    else:
        retorno = numeros[i] / numeros[i - 1] - 1
        retornos.append(retorno)

serie_retornos = pd.Series(retornos)
print(serie_retornos)
serie_retornos = serie_retornos.dropna()
serie_retornos = serie_retornos.replace([-1, np.inf], 0) #substituindo mais de um elemento de vez
print(serie_retornos)
```
