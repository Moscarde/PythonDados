# Pandas

```python
import pandas as pd
import yfinance as yf
```

```python
acoes= yf.download(['WEGE3.SA', 'VALE3.SA', 'PCAR3.SA', 'LREN3.SA',
        'PETR4.SA', 'EQTL3.SA', 'EGIE3.SA', 'ELET3.SA'])['Adj Close']
acoes
```

# Problemas
    - Base de dados - Cada coluna deve ter uma única variável
    - Agrupamento - É impossível agrupar dados de uma unica empresa utilizando o groupby
    - Filtragem - filtrar uma única empresa é possivel atravpes de colunas, nao por metodo de filtragem
    - Facilita também pro SQL


```python
# transformando data (index) para coluna
acoes  = acoes.reset_index()
acoes
```

```python
#melt ~> transforma um dataframe largo em um dataframe comprido
#cada coluna passa a ter uma única categoria: data, empresa e preço

dados_melted = pd.melt(acoes, id_vars = 'Date', var_name= 'empresa', value_name='price')
dados_melted
```

### Comando SQL
    - SELECT * FROM dados_cotacoes WHERE empresa = "WEGE3.SA"

```python
#fechamento das empresas
dados_melted.groupby('empresa')['price'].last()
```

```python
#preco médio
dados_melted.groupby('empresa')['price'].mean()
```

```python
#filtrar dados
dados_melted[dados_melted['empresa'] == 'WEGE3.SA']
```

```python
#Operacao inversa
#pivot_table
dados_pivoted = pd.pivot_table(dados_melted, index='Date', columns= 'empresa')
print(dados_pivoted.head(5))
#price vira um multindex
dados_pivoted = dados_pivoted.droplevel(level= 0, axis= 1)
print(dados_pivoted.head(5))
```

```python
#os dados assim são ótimos para aplicar funções de series temporais
dados_pivoted.pct_change() #'Valor de alteração entre dias'
```
