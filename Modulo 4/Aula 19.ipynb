# Pandas

```python
import pandas as pd
from datetime import datetime
import yfinance as yf
# import warnings

# warinings.filterwarnings('ignore')
pd.options.mode.chained_assignment = None
```

```python
dados = yf.download(tickers = ['ABEV3.SA', 'MGLU3.SA', 'WEGE3.SA', 'VALE3.SA'])['Adj Close']
dados
```

```python
#calcular a media de retorno diario
#retorno diario = diferença entre dia anterior e dia atual
media_retornos_diarios = dados.pct_change().dropna().mean()
media_retornos_diarios
```

```python
#usando map() para transformar dados de uma coluna
#media de retornos diarios anualizados
dados_anuais = media_retornos_diarios.map(lambda x: (1+x) ** 252 - 1)

dados_anuais
```

```python
print(type(dados_anuais))
dados_anuais = dados_anuais.to_frame()
dados_anuais
```

```python
dados_anuais = dados_anuais.reset_index()
print(dados_anuais)

dados_anuais.columns = ['nome_empresas', 'retornos_anuais']
dados_anuais
```

```python
#deixando nomes em caixa baixa ~~> map + lower
dados_anuais['nome_empresas'] = dados_anuais['nome_empresas'].map(lambda x: x.lower())
print(dados_anuais)
```

### Desafio

```python
#df base
dict_desafio = {
    "tickers" : ["WEGE3", "PETR3", "VALE3", "PETR4", "LREN3"],
    "cotacoes": [20, 30, 40, 12, 35],
    "nomes": ['Weg', 'Petro', 'Vale', 'Petro', 'Lojas Renner'],
    "preco_sobre_lucro": [25, 6, 12, 7, 25],
    "volume": [5000, 1000, 4000, 7000, 1200]
    }

df = pd.DataFrame(dict_desafio, index= [1, 2, 3, 4, 5])
df = df.set_index("tickers")
df.columns = ['preco', 'nomes', 'preco_sobre_lucro', 'volume']
df['lucro_por_acao'] = df['preco'] / df['preco_sobre_lucro']
dolar = 5.25
df['preco_em_dolar'] = df['preco'] / dolar
df['rank_lpa'] = df['lucro_por_acao'].rank(method = 'dense')
df
```

```python
df['nomes'] = df['nomes'].map(lambda x: x.lower())
df['preco'] = df['preco'].astype(str).map(lambda x: 'R$' + x)
# metodo gabarito
# df['preco'] = df['preco']..map(lambda x: 'R$' + str(x))
```
