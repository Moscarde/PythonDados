# Pandas

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
acao = 'WEGE3.SA'
dados = yf.download(tickers = acao)
dados
```

```python
#plotando um grafico
dados.plot(xlabel = 'Date', ylabel = 'Volume')
```

```python
#criando uma nova coluna com dados obtidos
dados['Volume Financeiro'] = dados['Volume'] * dados['Close']
dados
```

```python
# filtrar por range de datas

acao = 'PETR4.SA'

dados = yf.download(
    tickers = acao,
    start= '2010-01-01',
    end= datetime.now()
    )
dados
```

```python
# filtrar por range de datas
# e intervalos definidos, semanal, mensal, anual, trimestral,....

acao = 'PETR4.SA'

dados = yf.download(
    tickers = acao,
    start= '2010-01-01',
    end= datetime.now(),
    interval = '1wk'
    )

#existem muitos parametros para filtrar o resultado, porem Ã© bom ter uma boa
#amostra de dados
dados
```

```python
#puxando varias acoes simultaneas
#retorna um dataframe Multi Index
acoes = ['META', 'NKE', 'GOOGL', 'AMZN']

dados = yf.download(tickers = acoes, start='2015-01-01', end= datetime.now())

dados
```
