# MatplotLib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

<center><img src='figure.jpg' width=500>

```python
cotacao_weg = yf.download('WEGE3.SA')['Adj Close']
fig, ax = plt.subplots()
ax.plot(cotacao_weg.index, cotacao_weg.values)
```

```python
# Adicionando legendas nos eixos
# Adicionando titulo
fig, ax = plt.subplots()
ax.plot(cotacao_weg.index, cotacao_weg.values)
ax.set_xlabel('Data', color = 'b')
ax.set_ylabel('Cotacao', color = 'b', fontsize= 25, labelpad=10)
ax.set_title('WEGE3', fontweight = 'bold')
```

## Desafio

```python
acao = str(input('Digite uma ação:').upper()) + '.SA'
cotacao = yf.download(acao)['Adj Close']
cotacao = cotacao.resample('Y').mean()

fig, ax = plt.subplots()
ax.plot(cotacao.index, cotacao.values)
ax.grid(True)
ax.set_title(f'Historico de preço {acao} - Média Anual', fontweight = 'bold', fontsize = 15)
ax.set_xlabel('Data', labelpad= 15)
ax.set_ylabel('Valores em Reais', labelpad= 15)

```
