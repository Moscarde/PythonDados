# MatplotLib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timedelta
```

```python
#colocando o y em reais e na cor verde
# cotacoes = yf.download(['WEGE3.SA', 'PETR4.SA', 'VALE3.SA', 'ASAI3.SA'])['Adj Close']
cotacoes = yf.download(['WEGE3.SA'])['Adj Close']

fig, ax = plt.subplots()

ax.plot(cotacoes.index, cotacoes.values)
ax.tick_params(axis= 'both', colors='g')
ax.yaxis.set_major_formatter('R${x:1.2f}')
#'R${x:1.0f}' = R$40
#'R${x:1.2f}' = R$40,00
```

```python
#Mudando a ocorrencia das datas no eixo x
import matplotlib.dates as mdate

# cotacoes = cotacoes weg

fig, ax = plt.subplots()

ax.plot(cotacoes.index, cotacoes.values)
# intervalo de ticks

ax.xaxis.set_major_locator(mdate.YearLocator(5))
```

```python
# cotacoes = weg

fig, ax = plt.subplots()

ax.plot(cotacoes.index, cotacoes.values)

#guardando a formatacao em uma var
minha_formatacao = mdate.DateFormatter('%b-%Y')

#setando uma formatação para a exibição no eixo x
ax.xaxis.set_major_formatter(minha_formatacao)

ax.xaxis.set_major_locator(mdate.MonthLocator(interval= 10))

plt.xticks(rotation=75)
```

```python
#colocando o eixo em %
import matplotlib.ticker as mtick
# cotacoes = weg3

retornos = cotacoes[cotacoes.index.year > 2015].pct_change().dropna()

retornos_acumulados = (1+ retornos).cumprod() -1

fig, ax = plt.subplots()

ax.plot(retornos_acumulados.index, retornos_acumulados.values)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
#PercentFormatter() param 1 ~> valor já em valor inteiro (1=100%, 8=800%)
#param 0 ou sem param ~> valor está em % (1=1%, 10=10%)
```

```python
#mudando a frequencia dos ticks que nao sao datas

# cotacoes = weg3
retornos = cotacoes.pct_change().dropna()

retornos_acumulados = (1 + retornos).cumprod() -1

dias = range(0, len(retornos_acumulados))

fig, ax = plt.subplots()
ax.plot(dias, retornos_acumulados.values)
ax.set_title(f'Retornos acumulados WEG3 por dia (i={retornos_acumulados.index[0].date()})')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))

ax.xaxis.set_ticks(np.arange(0, len(retornos_acumulados), 800))
ax.xaxis.set_ticks([0, 2000, 4000], ['Inicio', 'Meio', 'Fim'])

print(np.arange(0, len(retornos_acumulados), 800))
```
