```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as mtick
import yfinance as yf
import numpy as np
```

```python
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
print(type(tips))

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
print(tips)
```

```python
female_tips = tips[tips['sex'] == 'Female']['tip'] 
male_tips = tips[tips['sex'] == 'Male']['tip']

smoker = tips[tips['smoker'] == 'Yes']
not_smoker = tips[tips['smoker'] == 'No']

not_smoker
```

```python
plt.figure(figsize=(20, 8))

plt.subplot(221)  # 2x2 grid, gráfico 1
sns.lineplot(data=male_tips)
plt.title("Male tips")

plt.subplot(222)  # 2x2 grid, gráfico 1
sns.lineplot(x = tips['total_bill'], y = tips['tip'])
plt.title("Tips / total billl")
# plt.subplot(112)

plt.subplot(223)
sns.histplot(data=tips['tip'], bins=15)
plt.title("Average tip value")

plt.subplot(224)
sns.histplot(data=not_smoker['tip'], bins=15, binwidth=1, label='Not Smoker')
sns.histplot(data=smoker['tip'], bins=15, binwidth=1, label='Smoker')
plt.title("Average tip value smoker and not smoker")
plt.legend()
plt.xlabel('Tip value')
```

```python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)

sns.set_theme()
sinplot(10, 1)
sinplot(4, 10)
sinplot(20, 20)
```
