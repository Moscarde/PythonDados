```python
cotacao_weg = yf.download('WEGE3.SA')['Adj Close']
fig, ax = plt.subplots()
ax.plot(cotacao_weg.index, cotacao_weg.values)
```

```python
# Adicionando legendas nos eixos
fig, ax = plt.subplots()
ax.plot(cotacao_weg.index, cotacao_weg.values)
ax.set_xlabel('Data', color = 'b')
ax.set_ylabel('Cotacao', color = 'b', fontsize= 25, labelpad=10)
```

```python
from datetime import datetime, timedelta
delta_ano = (datetime.now() - timedelta(days = 365)).date()

acao = str(input('Digite uma ao:').upper()) + '.SA'
cotacao = yf.download(acao, delta_ano)['Adj Close']

if cotacao[-1] > cotacao[0]:
    balanco = {'legenda': 'Cotao no tem r','color': 'g'}
else:
    balanco = {'legenda': 'Cotao afundando','color': 'r'}

fig, ax = plt.subplots()
ax.plot(cotacao.index, cotacao.values)
ax.grid(True)
ax.set_xlabel('Sou sardinha olhando curto prazo', labelpad= 15)
ax.set_ylabel(balanco['legenda'], color=balanco['color'], labelpad= 15)

```

```python
#poderia ser feito usando a metrica de rentabilidade
rentabilidade = (cotacao.iloc[-1] / cotacao.iloc[0]) - 1
print(rentabilidade > 0)
rentabilidade
```
