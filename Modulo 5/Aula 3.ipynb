# MatplotLib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
#parametros opcionais para globalizar
params = {
    'ytick.color': 'black',
    'xtick.color': 'black',
    'axes.labelcolor': 'black',
    'axes.edgecolor': 'black'
}
plt.rcParams.update(params)
```

```python
fig, ax = plt.subplots()

ax.grid(True)
ax.plot([2, 4, 6], [8, 5, 16])
```

```python
fig, ax = plt.subplots(2, 2, figsize = (20, 15))
for i in range(2):
    for j in range(2):
        print(i, j)
        ax[i, j].plot(np.random.random(10))
```

```python
cotacao_weg = yf.download('WEGE3.SA')['Adj Close']

fig, ax = plt.subplots()
ax.plot(cotacao_weg.index, cotacao_weg.values)

# ===
#cotacao_weg.plot()
```

```python
cotacoes = yf.download(['WEGE3.SA', 'PETR4.SA', 'VALE3.SA', 'MGLU3.SA'])['Adj Close'].dropna()

fig, ax = plt.subplots(2, 2, figsize = (20, 7), sharey= True) 
#sharey = compartilha o y ~> achata outros graficos, mantem um padrao

coluna = 0
for i in range(2):
    for j in range(2):
        print(i, j)
        ax[i, j].plot(cotacoes.index, cotacoes.iloc[:, coluna])
        ax[i, j].set_title(cotacoes.columns[coluna])
        coluna += 1
```

```python
acoes = []
for i in range(3):
    acoes.append(str(input('Digite o tick de uma acao').upper()) + '.SA')
acoes
```

```python
acoes.append('^BVSP')
dados_acoes = yf.download(acoes)['Adj Close'].dropna()

fig, ax = plt.subplots(2, 2, figsize = (12, 8))

coluna = 0
for i in range(2):
    for j in range(2):
        if dados_acoes.columns[coluna] == '^BVSP':
            ax[i, j].labelcolor = 'green'
        
        ax[i, j].plot(dados_acoes.index, dados_acoes.iloc[:, coluna])
        ax[i, j].set_title(dados_acoes.columns[coluna])
        coluna += 1
        
```
