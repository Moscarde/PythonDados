```python
import quantstats as qs
import yfinance as yf
import pandas as pd
```

```python
cotacoes = yf.download("WEGE3.SA", '2008-01-01')['Adj Close']
cotacoes
```

```python
retornos = cotacoes.resample('M').last().pct_change().dropna()

qs.extend_pandas()

retornos.plot_monthly_heatmap()
```

```python
retornos = cotacoes.resample('Y').last().pct_change().dropna()
print(retornos)
retornos.plot_yearly_returns()
```
