# MatplotLib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timedelta
```

<center><img src='figure.jpg' width=500>

```python
cotacoes = yf.download(['WEGE3.SA', 'PETR4.SA', 'VALE3.SA', 'ASAI3.SA'])['Adj Close']
```

```python
fig, ax = plt.subplots()

for acao in cotacoes:
    ax.plot(cotacoes[acao].index, cotacoes[acao].values, label = acao)
    # precisa passar o label em string no plot
ax.legend(loc = 0)
```

```python
acoes = []
for _ in range(4):
    acoes.append(str(
        input('Digite o nome de uma ação: '))
        .upper() + '.SA'
        )

acoes
```

```python
cotacoes = yf.download(acoes)['Adj Close']

fig, ax = plt.subplots()

for empresa in cotacoes:
    ax.plot(cotacoes.index, cotacoes[empresa].values, label = empresa)

ax.legend(loc = 0)

```
