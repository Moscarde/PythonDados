# MatplotLib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as mtick
import yfinance as yf
```

```python
# retorno acum ambev e bvsp 3anos 5anos 10anos

cotacoes = yf.download(['MULT3.SA', "^BVSP"],  '2010-01-01', '2023-01-01')['Adj Close']

retornos_mensais = cotacoes.resample('M').last().pct_change().dropna()
```

## Boxplot

```python
fig, ax = plt.subplots()

#plotando com o seaborn
sns.boxplot(data= retornos_mensais, orient = 'v')

#formatando com o matplot
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1, decimals = 0))
```

## Histograma

```python
fig, ax = plt.subplots()
sns.histplot(data= retornos_mensais['MULT3.SA'], bins=150)

ax.xaxis.set_major_formatter(mtick.PercentFormatter(1, decimals=0))
```

## Regress√£o linear

```python
retorno_diario = cotacoes.pct_change().dropna()

fig, ax = plt.subplots()

sns.regplot(x = retorno_diario['^BVSP'], y = retorno_diario['MULT3.SA'], line_kws={'color':'red'})

ax.xaxis.set_major_formatter(mtick.PercentFormatter(1, decimals= 0 ))
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1, decimals= 0 ))
```

## Grafico de linha padr√£o

```python
fig, ax = plt.subplots()

sns.lineplot(data= cotacoes)
ax.set_yscale('log')
print(cotacoes)
```

```python
cotacoes = yf.download(['PETR3.SA', 'PETR4.SA'])['Adj Close']

retornos_mensais = cotacoes.resample('M').last().pct_change().dropna()
retornos_anuais = cotacoes.resample('Y').last().pct_change().dropna()
```

```python
fig, ax = plt.subplots(2)

sns.regplot(x=retornos_mensais['PETR3.SA'],
    y= retornos_mensais['PETR4.SA'],
    line_kws={'color': 'red'}, 
    ax = ax[0])
ax[0].set_title('Regress√£o linear - mensal PETR3 + PETR4')

sns.regplot(x=retornos_anuais['PETR3.SA'],
    y= retornos_anuais['PETR4.SA'],
    line_kws={'color': 'red'}, 
    ax = ax[1])
ax[1].set_title('Regress√£o linear - anual PETR3 + PETR4')
```
