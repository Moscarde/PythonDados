## PLT x FIG, AX:
Existem duas formas de plotar graficos no matplotlib:
- Orientado a objeto(fig, ax)
- Estilo Pyplot (plt.plot())

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'x': np.random.random(10),
    'y': np.random.random(10),
})
df
```

```python
# forma orientado a objeto

fig, ax = plt.subplots()

ax.plot(df.index, df['x'])
ax.plot(df.index, df['y'])

ax.grid(True)
plt.show()
```

```python
#forma pyplot: figuras e eixos são criados automaticamenta
#nao é possivel configurar

plt.plot(df)
```
